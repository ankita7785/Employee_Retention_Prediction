A positive coefficient will indicate that an increase in the value of feature would increase the odds of the event occurring. On the other hand, a negative coefficient means the opposite, i.e,  an increase in the value of feature would decrease the odds of the event occurring.

### **7.3 Find the optimal cutoff** <font color = red>[15 marks]</font>

Find the optimal cutoff to improve model performance. While a default threshold of 0.5 was used for initial evaluation, optimising this threshold can enhance the model's performance.

First, plot the ROC curve and check AUC.



7.3.1 Plot ROC curve <font color="red">[3 Marks]</font>

```python
# Define ROC function
from sklearn.metrics import roc_curve,roc_auc_score
import matplotlib.pyplot as plt

def plot_roc_curve(y_true,y_pred_prob):
    fpr,tpr,threshold = roc_curve(y_true,y_pred_prob)
    auc_score =roc_auc_score(y_true,y_pred_prob) # aur score
    # plot curve
    plt.figure(figsize=(8,6))
    plt.plot(fpr,tpr,color='blue')
    plt.plot([0,1],[0,1],label='Random Classifier')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Recall)')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Call the ROC function
plot_roc_curve(df_predictions['Actual_Stayed'], df_predictions['Predicted_Probability'])
```

**Sensitivity and Specificity tradeoff**

Check sensitivity and specificity tradeoff to find the optimal cutoff point.

7.3.2 Predict on training set at various probability cutoffs <font color="red">[1 Mark]</font>

```python
# Predict on training data by creating columns with different probability cutoffs to explore the impact of cutoff on model performance
cutoffs= np.arange(0.0,1.05,0.05)
performance=[]
for cutoff in cutoffs:
    df_predictions['Predicted']=df_predictions['Predicted_Probability'].apply( lambda x: 1 if x > cutoff else 0 )

    accuracy = metrics.accuracy_score(df_predictions['Actual_Stayed'],df_predictions['Predicted'])
    recall = metrics.recall_score(df_predictions['Actual_Stayed'],df_predictions['Predicted'])
    precision = metrics.precision_score(df_predictions['Actual_Stayed'],df_predictions['Predicted'])

    performance.append([cutoff,accuracy,recall,precision])

cutoff=pd.DataFrame(performance,columns=['cutoff','accuracy','recall','precision'])

cutoff
```

7.3.3 Plot for accuracy, sensitivity, specificity at different probability cutoffs <font color="red">[2 Marks]</font>

```python
# Create a DataFrame to see the values of accuracy, sensitivity, and specificity at different values of probability cutoffs
cutoffs= np.arange(0.0,1.05,0.05)
metrics_list=[]
for cutoff in cutoffs:
    df_predictions['Predicted']=df_predictions['Predicted_Probability'].apply( lambda x: 1 if x > cutoff else 0 )

    cm= confusion_matrix(df_predictions['Actual_Stayed'],df_predictions['Predicted'])

    tn,fp,fn,tp=cm.ravel()
    
    accuracy = metrics.accuracy_score(df_predictions['Actual_Stayed'],df_predictions['Predicted'])
    sensitivity=tp/(tp+fn)
    specificity=tn/(tn+fp)

    metrics_list.append([cutoff,round(accuracy,3),round(sensitivity,3),round(specificity,3)])

cutoff_metrics_df =pd.DataFrame(metrics_list,columns=['cutoff','accuracy','sensitivity','specificity'])

cutoff_metrics_df
```

```python
# Plot accuracy, sensitivity, and specificity at different values of probability cutoffs
plt.figure(figsize=(10,6))
plt.plot(cutoff_metrics_df['cutoff'],cutoff_metrics_df['accuracy'],label='Accuracy')
plt.plot(cutoff_metrics_df['cutoff'],cutoff_metrics_df['sensitivity'],label='sensitivity')
plt.plot(cutoff_metrics_df['cutoff'],cutoff_metrics_df['specificity'],label='specificity')
plt.title('Accuracy Sensitivity and specificity of diff cutoff')
plt.xlabel('Probability cutoff')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```
